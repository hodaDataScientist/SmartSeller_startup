```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from bertopic import BERTopic
from nltk.corpus import stopwords
import nltk
```

```python
# Load your Persian documents
df = pd.read_excel('TopicModeling_sample2.xlsx')
df = pd.DataFrame(df)
df = df.dropna()


# Ensure your text data is in a list and convert all entries to strings
documents = df['Content'].astype(str).tolist()

# Create a SentenceTransformer model specifically for our language
model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')

# Initialize BERTopic model
topic_model = BERTopic(embedding_model=model)

# Fit the model on your documents
topics, probabilities = topic_model.fit_transform(documents)
```

```python
# Get the topic information
topic_info = topic_model.get_topic_info()
print(topic_info)
```

```python
len(df)
```

```python
# Display topics and top words for each topic
for topic in topic_model.get_topics().items():
    topic_num, words = topic
    print(f"Topic {topic_num}: {words}")
```

```python
# Visualize the topics
fig = topic_model.visualize_topics()
fig.show()
```

```python
# Optional: Visualize topic distributions
fig_dist = topic_model.visualize_documents(documents)
fig_dist.show()
```

```python
topic_6_sentences = [doc for i, doc in enumerate(documents) if topics[i] == 16]

# Print the sentences that belong to topic 6
for sentence in topic_6_sentences:
    print(sentence)
```
